```python
import yfinance as yf
import pandas as pd

tickers = ["BTC-USD", "CL=F"]
startDate = "2024-01-01"
endDate = "2024-12-31"

data = yf.download(tickers, start=startDate, end=endDate,)
display(data.head())
```

# HOW TO TURN A NOMINAL DATA INTO A RATIO DATA:

```python
# Select the 'Close' prices and calculate daily returns
closePrices = data['Close']
dailyReturns = closePrices.pct_change()*100

# Drop rows with NaN values
dailyReturns = dailyReturns.dropna()

# Display the daily returns
display(dailyReturns)
```

# Regression

```python
import statsmodels.api as sm

# Define the dependent and independent variables
Y = dailyReturns['BTC-USD']
X = dailyReturns['CL=F']

# Add a constant (Intercept) to the independent variable
X = sm.add_constant(X)

# Perform OLS Regression
model = sm.OLS(Y, X).fit()

# Display/Print the summary of the regression
print(model.summary())
```

# Predictive Model using Machine Learnng (i used gemini for this)

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np

# Prepare the data
# Assuming X and Y are already defined from the previous steps
# X is dailyReturns['CL=F'] and Y is dailyReturns['BTC-USD']

# Split the data into training and testing sets (80/20 split)
X_train, X_test, Y_train, Y_test = train_test_split(X['CL=F'], Y, test_size=0.2, random_state=42)

# Reshape X_train and X_test to be 2D arrays
X_train = X_train.values.reshape(-1, 1)
X_test = X_test.values.reshape(-1, 1)

# Initialize and train the Random Forest Regressor model
model = RandomForestRegressor(n_estimators=100, random_state=42) # You can adjust n_estimators
model.fit(X_train, Y_train)

# Make predictions on the test set
Y_pred = model.predict(X_test)

# Calculate RMSE and MAE
rmse = np.sqrt(mean_squared_error(Y_test, Y_pred))
mae = mean_absolute_error(Y_test, Y_pred)

# Display the results
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
```
